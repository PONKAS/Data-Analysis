##### import libraries

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
```

##### import datasets

```python
civ_marriages = pd.read_csv('Civil_Marriages.csv') #civil marriages dataset
civ_union = pd.read_csv('Civil_Unions.csv') #civil unions dataset
cust_marriages = pd.read_csv('Customary_Marriages.csv') #customary marriages dataset
divorces = pd.read_csv('Divorces.csv') # divorces dataset
```

***

## Datasets Cleaning 

#### Civil Marriages

```python
civ_marriages.info()
```

```python
#BridegroomAge and BrideAge dtype must be changed to numeric values instead of objects
civ_marriages['BridegroomAge'] = pd.to_numeric(civ_marriages['BridegroomAge'], errors = 'coerce')
civ_marriages['BrideAge'] = pd.to_numeric(civ_marriages['BrideAge'], errors = 'coerce')
civ_marriages.dtypes
```

* civil marriage dataset has been cleaned

#### Civil Unions

```python
civ_union.info()
```

* civil unions dataset is clean

#### Customary Marriages

```python
cust_marriages.info()
```

```python
#BridegroomAge and BrideAge dtype must be changed to numeric values instead of objects
cust_marriages['BridegroomAge'] = pd.to_numeric(cust_marriages['BridegroomAge'], errors = 'coerce')
cust_marriages['BrideAge'] = pd.to_numeric(cust_marriages['BrideAge'], errors = 'coerce')
cust_marriages.dtypes
```

* Customary Marriages dataset has been cleaned

#### Divorces

```python
divorces.info()
```

```python
#MaleAge and FemaleAge dtype must be changed to numeric values instead of objects
divorces['MaleAge'] = pd.to_numeric(divorces['MaleAge'], errors = 'coerce')
divorces['FemaleAge'] = pd.to_numeric(divorces['FemaleAge'], errors = 'coerce')
divorces['MarriageYear'] = pd.to_numeric(divorces['MarriageYear'], errors = 'coerce')
divorces.dtypes
```

```python
divorces.describe()
```

```python
divorces['DurationOfMarriage'].unique()
```

* 999 = cannot tell how long marriage lasted
* 0 = same year divorces

###### all datasets have been cleaned

# Data Analysis

## Civil Marriages Analysis

```python
sns.displot(civ_marriages, x="BridegroomAge", y="BrideAge",binwidth=(2, .5), cbar=True)
```

```python
sns.scatterplot(data=civ_marriages,x="BridegroomAge", y="BrideAge", hue = "ProvinceCode")
```

```python
g = sns.FacetGrid(civ_marriages, col="ProvinceCode", hue = "AgeDiff" ,col_wrap = 3)
g.map_dataframe(sns.scatterplot, x="BridegroomAge", y="BrideAge")
g.add_legend()
```

```python
h=sns.histplot(data=civ_marriages, x="ProvinceCode", hue="AgeDiff")
plt.xticks(rotation=70)
plt.figure(figsize=(15,8))
plt.tight_layout()
```

```python
ad = civ_marriages['AgeDiff'].value_counts()
ad.plot.pie(autopct="%.1f%%", figsize = (5,5))
```

```python
v1 = civ_marriages['BridegroomStatusCode'].value_counts()
v2 = civ_marriages['BrideStatusCode'].value_counts()
v1.plot.pie(autopct="%.1f%%", figsize = (10,10))
```

```python
v2.plot.pie(autopct="%.1f%%", figsize = (10,10))
```

```python
t = civ_marriages[civ_marriages['ProvinceCode'] != "Unspecified" ]
t = t[t['ProvinceCode'] != "Unknown"]
```

```python
t2=t['ProvinceCode'].value_counts()
t2
```

```python
t2.plot.pie(autopct="%.1f%%", title='Civil Marriages Registered in 2020', ylabel=' ' ,figsize = (10,10))
```
